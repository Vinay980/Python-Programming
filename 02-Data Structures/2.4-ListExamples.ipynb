### Real world examples Using Lists in Python

Lists are one of the most commonly used data structures in Python, thanks to their versatility and ease of use. Here are several practical examples that illustrate their use in real-world scenarios.

### Example 1. Manage a to-do List

--> Create a to-do List to Keep Tracking of Tasks

```python
to_do_list=["Buy groceries", "Clean the house", "Pay bills", "Call mom"]

#Add a new task to the list
to_do_list.append("Finish homework")
to_do_list.append("Go for a run")

#Remove a task from the list
to_do_list.remove("Pay bills")

#Check if a task is in the list
if "Call mom" in to_do_list:
    print("Don't forget to call mom!")
    
print("To-Do List:")
for task in to_do_list:
    print(f"- {task}")
```

### Example 2: Organizing Students Grades

Create a List to store and calculate average grades for students

```python
#Organize the students by their grades
grades = [85, 92, 78, 96, 88]

#Add a new grade to the list
grades.append(90)

#Calculate the average grade
average_grade = sum(grades) / len(grades)
print(f"\nAverage Grade: {average_grade:.2f}")

#find the highest and lowest grades
highest_grade = max(grades)
lowest_grade = min(grades)
print(f"Highest Grade: {highest_grade}")
print(f"Lowest Grade: {lowest_grade}")
```
