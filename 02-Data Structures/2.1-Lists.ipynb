List:-
01. Lists are ordered, mutable collections of items.
02. They can contain multiple data types

```python
list=[1, 2, 3,"four", "five", 6]
print(list)
```

```python
fruits=["apple", "banana", "cherry"]
print(fruits)
```

```python
# List can contain different data types
mixed_list=[1, "two", 3.0, [4, 5], {"six": 6}]
print(mixed_list)
```

```python
# list methods
my_list=[1, 2, 3]
my_list.append(4) ## Adds an element to the end of the list
print(my_list)
my_list.insert(0,0) ## Inserts an element at a specified position
print(my_list)
my_list.remove(0) ## Removes the first occurrence of a specified value
print(my_list)  
pop_list=my_list.pop() ## Removes and returns the last item in the list
print(pop_list) 
print(my_list)
my_list.clear() ## Removes all items from the list
```

```python
fruits=["apple", "elderberry", "grape", "cherry", "date", "fig","banana" ]
fruits.sort() ## Sorts the list in ascending order
print(fruits)
```

```python
numbers=[5, 2, 9, 1, 5, 6, 3, 4, 8, 7, 0]
print(numbers[1:5]) ## Slicing the list to get a sublist
print(numbers[::2]) ## Slicing the list to get every second element
print(numbers[1:5:2]) ## Slicing the list to get every second element from index 1 to 4
```

```python
# Iterating through a list
numbers=[5, 2, 9, 1, 5, 6, 3, 4, 8, 7, 0]
for number in numbers:
    print(number)
```

```python
# Iterating with index
numbers=[5, 2, 9, 1, 5, 6, 3, 4, 8, 7, 0]
for index,numbers in enumerate(numbers):
    print(f"Index: {index}, Number: {numbers}")
```

```python
# List comprehension
lst = []
for i in range(10):
    lst.append(i**2)
print(lst)

# OR
[i**2 for i in range(10)]
```

### List Comprehension 

Basic Syntax  [expression for item in iterable]

with condition logic  [expression for item in iterable if condition]

```python
### List Comprehension with condition
lst = []
for i in range(10):
    if i % 2 == 0:
        lst.append(i**2)
print(lst)
```

```python
list=[i**2 for i in range(10) if i % 2 == 0]
print(list)
```

```python
lst1=[1, 2, 3]
lst2=['a', 'b', 'c']
pair=[(i,j) for i in lst1 for j in lst2]
print(pair)
```

```python
## list of comprehension with function calls
words = ["hello", "world", "python", "programming"]
length=[len(word) for word in words]
print(length)
```
