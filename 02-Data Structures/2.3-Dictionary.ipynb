### Dictionary

Dictionaries are unordered collections of items. They store data in key-value pairs. Keys must be unique and immutable (eg.-> strings, numbers, or tuples), while values can be of any type.

```python
empty_dist={}
print(type(empty_dist))
```

```python
empty_dist = dict()
print(type(empty_dist))
```

```python
# single key-value pair is always used to create a dictionary
student  = {"name":"Vinay", "age":28, "course":"Python"}
print(student)
print(type(student))
```

```python
#accessing values from a dictionary
print(student["name"])
print(student["course"])
print(student["age"])
```

```python
#accessing values from a dictionary using get() method
print(student.get("name"))
print(student.get("course"))
```

```python
#modifying values in a dictionary
student["age"] = 29
student["address"] = "Bangalore"   #update a new data in the dictionary
print(student)
```

```python
# deleting a key-value pair from a dictionary
del student["age"]
print(student)
```

```python
# shallow copy of a dictionary
student_copy = student.copy()
print(student_copy)
```

```python
#itrating through a dictionary
for key in student:
    print(key)
```

```python
#iterating through a dictionary using items() method
for key, value in student.items():
    print(f"{key}:{value}")
```

```python
#iterating over values in a dictionary
for value in student.values():
    print(value)
```
